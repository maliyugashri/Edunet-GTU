7. if-elif-else Statement

```python
print("\n----if-elif-else Statement----")
marks = 80

if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
else:
    print("Grade: D")
```

8. Nested id-else


```python
print("\n----Nested if----")
number = 10
if number > 0:
    if number % 2 == 0:
        print("Number is positive and even")
    else:
        print("Number is positive and odd")
else:
    print("Number is negative")
```

9. Break, Continue , Pass Example

```python
print("\n----Break Example----")
for i in range(1, 11):
    if i == 5:
        break
    print(i)

print("\n---continue---")
for i in range(1, 6):
  if i == 2:
    continue
  print(i)

print("\n---pass---")
for i in range(1, 4):
  if i == 3:
    pass
  else:
    print(i)
```
